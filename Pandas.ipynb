```python
import pandas as pd

# Create long-format student data
df = pd.DataFrame({
    'Student': ['Alice', 'Alice', 'Bob', 'Bob', 'Charlie', 'Charlie',
                'Alice', 'Alice', 'Bob', 'Bob', 'Charlie', 'Charlie'],
    'Subject': ['Math', 'Science'] * 6,
    'Year': ['2023'] * 6 + ['2024'] * 6,
    'Marks': [90, 88, 85, 80, 95, 92, 94, 90, 89, 82, 96, 94]
})
```

```python
df
```

```python
grouped = df.groupby(['Student', 'Subject'])
```

```python
print(grouped.size())
```

```python
print(grouped.agg(['max','min']))
```

```python
df['Prev_Marks'] = grouped['Marks'].shift(1)
```

```python
df
```

```python
df['Rank_in_Subject'] = df.groupby('Subject')['Marks'].rank(ascending=False, method='dense')
```

```python
df
```

```python
df['Cumulative_Marks'] = grouped['Marks'].cumsum()
```

```python
df
```

```python
df['CumMax'] = grouped['Marks'].cummax()
df['CumMin'] = grouped['Marks'].cummin()
df['CumProd'] = grouped['Marks'].cumprod()
```

```python
df
```
